```
import keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
dir = ["left","right","straight","up"]
mood = ["angry","happy","neutral","sad"]
glasses = ["open","sunglasses"]
X = []
y = []
```

```
for i in dir:
  for j in mood:
    for k in glasses:
      for l in range(1,21):
        try:
          img = cv2.imread("/content/drive/My Drive/Advanced_Assignment_Dataset/"+ f"{i}_{j}_{k}_{l}"+ ".pgm")
          arr = keras.preprocessing.image.img_to_array(img)
          X.append(arr)
          y.append([i,j,k])
          print( f"{i}_{j}_{k}_{l}")
        except:
          print("error detected")
          pass
```

```
X = np.array(X)
y = np.array(y)
np.shape(X),np.shape(y)
```

```
X = X.astype('float')
X = X / 255
```

```
np.save("/content/drive/My Drive/Dataset_Advanced_Features.npy",X)
np.save("/content/drive/My Drive/Dataset_Advanced_Label.npy",y)
```

```
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(y)
y = enc.transform(y).toarray()
```

```
enc.get_feature_names(['dir', 'mood','glasses'])
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras import regularizers
from keras.layers import Dropout
```

```
model = Sequential()
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(80,activation = 'relu'))
model.add(Dense(80,activation = 'relu'))
model.add(Dense(10,activation = 'sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics = ['accuracy'])
model.fit(X_train,y_train,epochs=15,validation_split=0.2)
```

```
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(10, activation='softmax'))


model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])

model.fit(X_train,y_train,epochs=15,validation_split=0.2)
```

```
y_test_pred = model.predict(X_test)
print(y_test_pred)
```

```
y_test_pred = enc.inverse_transform(y_test_pred)
enc_2 = OneHotEncoder(handle_unknown='ignore')
enc_2.fit(y_test_pred)
y_test_pred = enc_2.transform(y_test_pred).toarray()
enc_2.get_feature_names(['dir', 'mood', 'glasses'])
```

```
model.evaluate(X_test,y_test)
```

```
from sklearn.metrics import confusion_matrix,accuracy_score

acc = accuracy_score(y_test,y_test_pred)
acc
```
